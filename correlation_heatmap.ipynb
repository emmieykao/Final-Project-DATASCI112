```python
import pandas as pd
df = pd.read_csv("datasci112_final.csv")
```

```python
#for the sake of readability
df_renamed = df.rename(columns={
    "Convenience stores/1,000 population": "Convenience Store Density",
    "Fast-food restaurants/1,000 population": "Fast Food Density",
    "% Household food insecurity": "Household Food Insecurity",
    "Grocery stores/1,000 population": "Grocery Store Density",
    "Median household income": "Median Income",
    "% Households, no car & low access to store": "No Car & Low Food Access",
    "% Population low income & low access to store": "Low Income & Low Access",
    "% Population with low access to store": "Low Food Access",
    "SNAP-authorized stores/1,000 population": "SNAP Store Density",
    "Supercenters/1,000 population": "Supercenter Density",
    "% Adults with Obesity": "Adult Obesity Rate",
    "% Adults with Diabetes": "Adult Diabetes Rate",
})
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
cols = [
    "Convenience Store Density",
    "Fast Food Density",
    "Grocery Store Density",
    "Median Income",
    "No Car & Low Food Access",
    "Low Income & Low Access",
    "Low Food Access",
    "SNAP Store Density",
    "Supercenter Density",
    "Retail Food Environment Index",
    "Adult Obesity Rate",
    "Adult Diabetes Rate",
    "Cancer Incidence Rate"]
corr = df_renamed[cols].corr()
plt.figure(figsize=(9,7))
sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    vmin=-1,
    vmax=1)
plt.title("Correlation Between Food Environment, Income, and Health Outcomes")
plt.show()
```

```python
predictors = [
    "Fast Food Density",
    "Low Food Access",
    "Median Income"]
outcomes = [
    "Adult Obesity Rate",
    "Adult Diabetes Rate"]
corr = df_renamed[predictors + outcomes].corr()
corr_subset = corr.loc[outcomes, predictors]
plt.figure(figsize=(8,4))
sns.heatmap(
    corr_subset,
    annot=True,
    cmap="coolwarm",
    vmin=-1,
    vmax=1,
    fmt=".2f",
    linewidths=0.5)
plt.title("Correlation of Food Environment and Income with Health Outcomes")
plt.show()
```
